```python
from pathlib import Path
import os,sys
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import math
formated_dir = Path("/data02/starmage/datasets/cc/formated/008")
files = os.listdir(formated_dir)
files.sort()
chosed_id = 0
```

```python
samples = 9
rows = int(math.sqrt(samples))
idx = [random.randint(0, len(files)-1) for _ in range(samples)]
fig, axs = plt.subplots(rows, rows, figsize=(16, 16))

for i in range(rows):
    for j in range(rows):        
        im= files[idx[i*rows+j]]  
        axs[i, j].imshow( mpimg.imread(formated_dir / f"{im}/image.png"))
        axs[i, j].axis("off")
```

```python
tg = formated_dir / f"{files[idx[0]]}/image.png"
bad_dir = "/data03/starmage/projects/UICoder/data/bad/"
os.system(f"cp {str(tg)} {bad_dir}/{chosed_id}.png")
chosed_id += 1
```
