```python
from datasets import load_dataset
from IPython.display import display
from PIL import Image as PILImage

from ...configs.paths import DATA_DIR


# Load parquet file
ds = load_dataset(
    "parquet",
    data_files=DATA_DIR / "datasets" / "all_samples.parquet",
    split="train"
)


sample = ds[30]

for key, value in sample.items():
    if key in ["root", "image_refs"]:
        print(f"{key}: {type(value)}")
        # root is a single image
        if key == "root" and value is not None:
            if isinstance(value, str):
                display(PILImage.open(value))
            else:
                display(value)
        # image_refs is a list
        elif key == "image_refs":
            for img in value:
                if isinstance(img, str):
                    display(PILImage.open(img))
                else:
                    display(img)
    else:
        print(f"{key}: {type(value)} -> {value}")
```
